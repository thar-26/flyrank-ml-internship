## 2. The question: decision, action, cost of a wrong call

The decision is which pages should receive limited review time first. A content editor or SEO reviewer would use the ranked queue to inspect high-priority pages and then decide whether to refresh, expand, protect, monitor, or take no action. A wrong recommendation could waste reviewer time on a page that does not need attention, while a missed opportunity could leave an important declining page unnoticed. Data and ML may help because the decision depends on several interacting signals, such as visibility, content age, position, CTR, and engagement, which may be difficult to combine effectively with a single hand-written rule.

## 3. Quick look at the data (2-3 real numbers)

The starter dataset contains 30,000 pages, and 54.2% are marked as declining under the starter proxy label. There are also 16,726 pages with at least 500 impressions, showing that many pages have enough visibility to make prioritization practically relevant.
